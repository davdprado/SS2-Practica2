```python
import pandas as pd
df = pd.read_csv('datos2.csv')
```

### Eliminar elementos filas duplicadas

```python

import re
dataLimpia = df.drop_duplicates()
dataLimpia['Rating'].fillna(0, inplace=True)
dataLimpia['Level'].fillna('Beginner level', inplace=True)
dataLimpia['Duration'].fillna('0 hour', inplace=True)
dataLimpia['Review'].fillna('0 reviews', inplace=True)
dataLimpia['Review'] = dataLimpia['Review'].apply(lambda x: int(re.findall(r'\d+', x)[0]))
dataLimpia['Duration'] = dataLimpia['Duration'].str.extract('(\d+)', expand=False).astype(float)
dataLimpia
```

_Limpiamos la Data donde el raiting, level y duracion cambiamos los datos para poder tener una mejor vision de los datos numericos_

### Calcular el Promedio
_cálculo del promedio de calificaciones para cada curso en el
conjunto de datos. Para ello, agrupen los datos por el título del curso y
calculen el promedio de las calificaciones para cada grupo._


```python
promediopc = dataLimpia.groupby('Course Title')['Rating'].mean()
promediopc
```

### Cursos con mayor y menor rating

```python
imax = dataLimpia['Rating'].idxmax()
imin = dataLimpia['Rating'].idxmin()

cmax = dataLimpia.loc[imax, 'Course Title']
rmax = dataLimpia.loc[imax, 'Rating']

cmin = dataLimpia.loc[imin, 'Course Title']
rmin = dataLimpia.loc[imin, 'Rating']

print('El curso con la calificación más alta es:', cmax, 'con una calificación de:', rmax)
print('El curso con la calificación más baja es:', cmin, 'con una calificación de:', rmin)
```

### Porcentaje de Cursos con horario flexible

```python
totalPercent = len(dataLimpia)
flexibleSchedule = dataLimpia[dataLimpia['Schedule'] == 'Flexible schedule']
cant = len(flexibleSchedule)
percent = (cant/totalPercent)*100
percent = round(percent, 2)
print('El porcentaje de cursos con horario flexible es: ' + str(percent)+'%')
```

## Graficos

### Cantidad de Cursos por nivel

```python
import matplotlib.pyplot as plt

course_per_level = dataLimpia['Level'].value_counts()

plt.figure(figsize=(10,5))
course_per_level.plot(kind='bar', color='skyblue')
plt.title('No. Cursos por nivel')
plt.xlabel('Nivel')
plt.ylabel('No. Cursos')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

_Esta grafica de barras muestra que hay muchos cursos para principiantes pues es ahi donde se enfoca la mayoria de cursos para despues pasar a los siguientes niveles, como resultados podemos definir que hay mas de 6000 cursos para principiantes, aproximadamente mas de 2000 de nivel intermedio y menos de 1000 para niveles avanzados, pues estos ultimos son los que requieren de mas dedicacion y estudio._

### Grafica numero de cursos por categorias (Keyword)

```python
top_categories = dataLimpia['Keyword'].value_counts().head(10)

plt.figure(figsize=(15,6))
top_categories.plot(kind='barh', color='red')
plt.title('Top 10 Categorias')
plt.xlabel('No. Cursos')
plt.ylabel('Categoria')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

_Con el grafico anterior definimos que los cursos de Social Sciences son los que mas existen seguidos de Desarrollo personal._
_En fin podemos ver que los cursos de Arts and Humanities lograron entrar a un top 10 con aproximadamente unos 500 cursos_

### Dispersion de visualizacion del curso

```python
plt.figure(figsize=(15,6))
scatter = plt.scatter(dataLimpia['Duration'], dataLimpia['Review'], c=dataLimpia['Review'], cmap='RdYlBu', alpha=0.6)
plt.title('Duracion vs Reviews')
plt.xlabel('Duracion')
plt.ylabel('Reviews')
plt.tight_layout()
plt.show()
```

_La grafica de dispersion muestra la relacion entre la duracion y reviews de cursos donde la concentracion de las reviews son en los cursos que tienen una duracion mas corta. aunque se puede observar que hay cursos con una duracion mayor a las 100 horas que cuentan con reviews apesar de que son de los cursos que menos escoge la gente._

### Histograma Duracion de los cursos

```python
plt.figure(figsize=(10,6))
plt.hist(dataLimpia['Duration'], bins=20, color='green', edgecolor='black')
plt.title('Histograma Duración de cursos')
plt.xlabel('Duración')
plt.ylabel('No. Cursos')
plt.grid(True)
plt.show()
```

_Con el histograma podemos se puede observar que muchos cursos estan en maximo 20 horas de duracion que es lo mas comun, pero si observamos bien hay un par que superan las 120 horas. Con esto determinamos que la mayoria de personas prefieren un curso con menos duracion en horas ya que es mucho mas facil y flexible optar por un curso de este tipo._

### Grafico de Cajas de distribuciones de calificaciones de los cursos por nivel de dificultad

```python
levels = dataLimpia['Level'].unique()
fig, ax = plt.subplots(figsize=(10, 6))

for level in levels:
    calificaciones = dataLimpia.loc[dataLimpia['Level'] == level, 'Rating']
    ax.boxplot(calificaciones, positions=[levels.tolist().index(level)], widths=0.6, patch_artist=True, boxprops=dict(facecolor='skyblue'))
ax.set_xticks(range(len(levels)))
ax.set_xticklabels(levels)
ax.set_title('Boxplot de calificaciones por nivel')
ax.set_xlabel('Nivel')
ax.set_ylabel('Calificación')
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

_En el grafico de cajas podemos observar que en los 3 niveles de dificultad que hay en nuestros tablas la concentracion de los datos estan en un rating de entre 4 a 5. Lo que quiere decir que hay una tendencia de que la mayoria de cursos tengan notas altas. Aunque podemos observar que siempre hay datos muy alejados de este y preocupa que hay Calificaciones en 0. Aunque no es una mayoria es curioso que almenos tengan una calificacion baja en estos niveles._
